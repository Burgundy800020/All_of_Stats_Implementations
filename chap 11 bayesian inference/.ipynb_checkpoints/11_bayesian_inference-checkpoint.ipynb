```python
# Chapter 11
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as scp
import math
```

```python
mu = 5
n = 100
xs = np.random.normal(mu, 1, n)
```

```python
#b Finding posterior
```

```python
#analytic
```

```python
xBar = np.mean(xs)
se = 1/n**0.5
```

```python
xcoords = np.linspace(xBar-4*se, xBar+4*se, 1000)
ycoords = scp.norm.pdf(xcoords, loc=xBar, scale=1/n**0.5)
```

```python
#simulation
```

```python
B = 1000
xSim = np.random.normal(xBar, se, B)
```

```python
plt.title("posterior distribution")
plt.subplot(1, 2, 1)
plt.plot(xcoords, ycoords)
plt.subplot(1, 2, 2)
plt.hist(xSim, bins = 20)
plt.show()
```

```python
#d find posterior for theta = e^{mu}
```

```python
#analytic
```

```python
theta = np.exp(xBar)
```

```python
def pdf_theta(theta):
    return np.exp(-n/2*(np.log(theta)-xBar)**2)/theta*(n/2/math.pi)**0.5
```

```python
xcoords = np.linspace(theta-50, theta+50, 1000)
ycoords = pdf_theta(xcoords)
```

```python
#simulation
```

```python
thetaSim = np.exp(xSim)
```

```python
plt.title("posterior distribution for e^{mu}")
plt.subplot(1, 2, 1)
plt.plot(xcoords, ycoords)
plt.subplot(1, 2, 2)
plt.hist(thetaSim, bins = 20)
plt.show()
```

```python
s1 = 30
n1 = 50
p1 = s1/n1
s2 = 40
n2 = 50
p2 = s2/n2
tau = p2-p1
tau
```

```python
#(a) estimate tau
```

```python
#delta method
```

```python
se = (p1*(1-p1)/n1 + p2*(1-p2)/n2)**0.5
se
```

```python
#bootstrap
```

```python
B = 10000
ps1 = np.random.binomial(n=n1, p=p1, size=B)/n1
ps2 = np.random.binomial(n=n2, p=p2, size=B)/n2
np.std(ps2-ps1)
```

```python
#bayesian, simulation
```

```python
ps1 = np.random.beta(s1+1, n1-s1+1, size=B)
ps2 = np.random.beta(s2+1, n2-s2+1, size=B)
taus = ps2-ps1
tauBay = np.mean(taus)
seBay = np.std(taus)
print(tauBay, seBay)
```

```python
#(d) estimate phi
```

```python
#delta method
```

```python
phi = np.log(p1*(1-p2)/p2/(1-p1))
phi
```

```python
se = (1/(n1*p1*(1-p1)) + 1/(n2*p2*(1-p2)))**0.5
se
```

```python
#Bayesian, simulation
```

```python
B = 10000
ps1 = np.random.beta(s1+1, n1-s1+1, size=B)
ps2 = np.random.beta(s2+1, n2-s2+1, size=B)
phis = np.log(ps1*(1-ps2)/ps2/(1-ps1))
phiBay = np.mean(phis)
seBay = np.std(phis)
print(phiBay, seBay)
```

```python
b = 1
x = 0.5
```

```python
#wald
W = np.abs(x)
p_value = 2*(1-scp.norm.cdf(W))
print(f"Wald test p value: {p_value}")
#bayesian testing
P0 = np.exp(-x**2/2)/(b**2+1)**(1.5)
print(f"Bayesian testing P(H0 | x): {P0}")
```
